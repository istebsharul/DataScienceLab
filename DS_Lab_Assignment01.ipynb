```python
text=open("sample.txt","r")

d=dict()
type=0
token=0

for line in text:
  #converting paragraph into line
  line=line.strip()

  # converting into lowercase
  line=line.lower()
  
  # Splitting the line into words
  words=line.split(" ")

  for word in words:
    token+=1
    if word in d:
      d[word]=d[word]+1
    else:
      d[word]=1
      type+=1

sorted_dict= {k:v for k, v in sorted(d.items(), key= lambda item: item[1], reverse=True)}

top_5= list(sorted_dict.items())[:-5]
print(top_5)
```

```python
print(type)
print(token)

ttr=type/token
print(ttr)
```

```python
# import nltk
import matplotlib.pyplot as plt
plt.figure(figsize=(30,10))
plt.bar(d.keys(), d.values())
plt.xlabel('Word')
plt.ylabel('Frequency')
# plt.title('Word Frequency Distribution')
plt.show()
```

```python
import nltk
from nltk import bigrams
from collections import Counter

# Generate the bigrams
bigrams = bigrams(text)

# Count the frequencies of the bigrams
bigram_freq = Counter(bigrams)

# Print the 10 most common bigrams
print(bigram_freq.most_common(10))
```
